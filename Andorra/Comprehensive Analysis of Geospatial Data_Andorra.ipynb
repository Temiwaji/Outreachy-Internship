#### 1. INTRODUCTION
    1.0 OVERVIEW

Andorra, a small and landlocked European coprincipality, is nestled among the southern peaks of the Pyrenees Mountains. It is bordered by France to the north and east, and by Spain to the south and west, making it one of the smallest states in Europe. The capital, Andorra la Vella, holds the distinction of being the highest capital city in Europe, situated at an elevation of 1,023 meters above sea level. Globally, Andorra ranks as the 16th-smallest country by land area and the 11th-smallest by population. Despite its size, Andorra attracts around 10.2 million visitors annually, a testament to its popularity as a tourist destination. It's noteworthy that Andorra is not a member state of the European Union but has been a proud member of the United Nations since 1993.

This project involves carrying out geospatial analysis to explore and understand the geographic context of the land sector data for Andorra (AND). The project's processes include;
